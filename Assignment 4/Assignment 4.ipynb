(d)	The products in the sales data can be further grouped as follow:

    a.	Category A: facecream, facewash, moisturizer
    b.	Category B: toothpaste
    c.	Category C: bathingsoap, shampoo

Group the data accordingly and obtain the following information, then plot into two subplots: 

    1.	total profit of each category for every month
    2.	percentage of each category for every month over the grand total profit of all categories.



```python
mapping = {'month_number':'month','facecream': 'A', 'facewash': 'A', 'toothpaste': 'B', 'bathingsoap': 'C', 'shampoo': 'C', 'moisturizer' : 'A'}
categories = sales.groupby(mapping, axis=1)
df = categories.sum().reset_index(drop=True)
df['A'] = df['A']*10
df['B'] = df['B']*10
df['C'] = df['C']*10
df['Percentage_A']=(df['A']/sales['total_profit'])*100
df['Percentage_B']=(df['B']/sales['total_profit'])*100
df['Percentage_C']=(df['C']/sales['total_profit'])*100

fig, axs = plt.subplots(2,figsize=(12,6), dpi=100)
fig.suptitle('Category subplots')
l1,=axs[0].plot(df['month'], df['A'], color='red',lw=3,marker='o', markersize=8, markerfacecolor="red")
l2,=axs[1].plot(df['month'], df['Percentage_A'], color='red',lw=3,marker='*', markersize=8, markerfacecolor="red")
l3,=axs[0].plot(df['month'], df['B'], color='blue',lw=3,marker='o', markersize=8, markerfacecolor="blue")
l4,=axs[1].plot(df['month'], df['Percentage_B'], color='blue',lw=3,marker='*', markersize=8, markerfacecolor="blue")
l5,=axs[0].plot(df['month'], df['C'], color='yellow',lw=3,marker='o', markersize=8, markerfacecolor="yellow")
l6,=axs[1].plot(df['month'], df['Percentage_C'], color='yellow',lw=3,marker='*', markersize=8, markerfacecolor="yellow")
axs[1].legend(handles = [l1,l2,l3,l4,l5,l6] , labels=['A', 'A', 'B','B', 'C', 'C'],loc='upper center', 
             bbox_to_anchor=(0.5, -0.2),fancybox=False, shadow=False, ncol=3)
axs[1].set_xlabel('Month')
axs[0].set_ylabel('Total Profit of Each Month')
axs[1].set_ylabel('Percentage over Total Profit')
plt.show()
```

```python
newdf = pd.merge(sale, product)
newdf['sales']=newdf['qty']*newdf['price']
qty_sales = newdf['qty'].groupby(sale['product']).sum().reset_index()
total_sales = newdf['sales'].groupby(sale['product']).sum().reset_index()

plt.plot(qty_sales['product'], qty_sales['qty'], color='blue',lw=3,marker='o', markersize=8, markerfacecolor="blue")
plt.bar(total_sales['product'], total_sales['sales'], color='red')
plt.show()
```
