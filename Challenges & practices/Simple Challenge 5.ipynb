```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('telco-customer-churn.csv')
```

```python
def diff(arr):
    return arr.max() - arr.min()

grouped = df['MonthlyCharges'].groupby([df['InternetService'],df['gender']])
grouped.agg(['mean', 'min','max',diff])
```

```python
grouped = df['OnlineSecurity'].groupby([df['SeniorCitizen'],df['OnlineSecurity']])
grouped.agg(['count'])
```

```python
def diff(arr):
    return arr.max() - arr.min()

grouped = df['tenure'].groupby([df['Churn'],df['InternetService']])
grouped.agg(['mean', 'min','max',diff])
```
