**PANDAS: TASK 1**

```python
import pandas as pd

emp_detail = pd.read_csv('emp_detail.csv')
```

**PANDAS: TASK 2**

```python
emp_detail.head()
```

**PANDAS: TASK 3**

```python
attendance = pd.DataFrame(att_arr)
emp_attendance = pd.concat([emp_detail, attendance], axis=1)
emp_attendance
```

**PANDAS: TASK 4**

```python
total_attd = emp_attendance.sum(axis=1)
emp_attendance['Total']=total_attd
emp_attendance
```

**PANDAS: TASK 5**

```python
attendance_list = emp_attendance.Total.tolist()
percent_list = []
for i in range(len(attendance_list)):
    percent = round((attendance_list[i]/14*100),2)
    percent_list.append(percent)
emp_attendance['Rate'] = percent_list 
emp_attendance
```

**PANDAS: TASK 6**

```python
rate_list = emp_attendance.Rate.tolist()
status_list = []
for i in range(len(rate_list)):
    if rate_list[i]>90:
        status_list.append('Good')
    elif rate_list[i]<90 and rate_list[i]>60 :
        status_list.append('Moderate')
    elif rate_list[i]<60 :
        status_list.append('Bad')
    else:
        print("Error!")
        
emp_attendance['Status'] = status_list 
emp_attendance
```

**PANDAS: TASK 7**

```python
def department_attendance(emp_attendance):
    fi_attendance,st_attendance = 0,0 
    
    for ind in emp_attendance.index : 
        if emp_attendance['Department'][ind]=='FINANCE':
            fi_attendance = fi_attendance + emp_attendance['Total'][ind]
        elif emp_attendance['Department'][ind]=='STAFFING':
            st_attendance = st_attendance + emp_attendance['Total'][ind]
            
    fi_attendance = fi_attendance/(emp_attendance.Department == 'FINANCE').sum()
    st_attendance = st_attendance/(emp_attendance.Department == 'STAFFING').sum()
    
    return fi_attendance,st_attendance;
```

```python
fi_attendance,st_attendance = department_attendance(emp_attendance)
print("Finance Department Average Attendance = ",fi_attendance)
print("Staffing Department Average Attendance = ",st_attendance)
```
